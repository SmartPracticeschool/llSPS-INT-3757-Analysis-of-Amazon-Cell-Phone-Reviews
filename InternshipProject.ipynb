#                ANALYSIS OF AMAZON CELL PHONE REVIEWS

```python
#Importing the libraries
import numpy as np
import pandas as pd
```

```python
#Importing the dataset
dataset = pd.read_csv(r'C:\Users\Deepika\Documents\20191226-reviews.txt', delimiter = '\t', quoting = 3)
```

```python
dataset
```

```python
#Checking for null values
dataset.isnull().any()
```

```python
#Replacing null values
dataset['title'].mode()
```

```python
dataset.fillna(dataset['title'].mode()[0],inplace = True)
```

```python
#Rechecking for null values
dataset.isnull().any()
```

```python
#Importing NLP libraries
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
#Applying NLP
c = []
for i in range(0,67986):
    review = dataset["title"][i]
    review = re.sub('[^a-zA-Z]',' ',review)
    review = review.lower()
    review = review.split()
    review = [word for word in review if not word in set(stopwords.words('english'))]
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    c.append(review)
```

```python
c
```

```python
#Importing libraries
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import MinMaxScaler
```

```python
cv = CountVectorizer(max_features = 1500)
ms = MinMaxScaler()
```

```python
cv
```

```python
x = cv.fit_transform(c).toarray()
```

```python
x.shape
```

```python
y = dataset.iloc[:,2:3].values
```

```python
y
```

```python
y = ms.fit_transform(y)
```

```python
y
```

```python
#Spitting data into train and test sets
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20, random_state = 0)
```

```python
#Importing libraries
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Initialising the model
model = Sequential()
```

```python
#Adding input layer
model.add(Dense(input_dim = 1500, kernel_initializer = "random_uniform", activation = 'relu', units = 1000))
```

```python
#Adding hidden layer
model.add(Dense(units = 750, kernel_initializer = "random_uniform", activation = 'relu'))
```

```python
#Adding output layer
model.add(Dense(units = 1, kernel_initializer = "random_uniform", activation = 'sigmoid'))
```

```python
#Configuring the learning process
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#Training the model
model.fit(x_train, y_train, epochs = 1, batch_size = 500)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_pred = (y_pred>0.5)
```
